```python
Health_dummies= pd.get_dummies(HealthData)
```

```python
X= Health_dummies.drop(columns= "Value")
Y= Health_dummies["Value"]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
train_x, val_x, train_y, val_y= train_test_split(X,Y, test_size= 0.4, random_state= 100)
lr= LinearRegression()
lr.fit(train_x,train_y)
train_pred= lr.predict(train_x)
test_pred= lr.predict(val_x)
```

```python
lr.coef_
```

```python
print("test_mse:",mean_squared_error(test_pred,val_y))
print("train_mse:", mean_squared_error(train_pred,train_y))
print("mae:", mean_absolute_error(test_pred,val_y))
print("test_r2:", r2_score(test_pred,val_y))
print("train_r2:", r2_score(train_pred,train_y))
```
